```python
import math

def func(inter, num, power):
    return inter ** power - num

def bisection(num, power, tol):
    a = 0
    b = max(1, num)
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if abs(func(c,num,power)) == 0:
            return c
        if abs(func(c, num,power)) < tol:
            return c
        elif func(c, num,power) * func(a, num,power) < 0:
            b = c
        else:
            a = c
    return c

print(bisection(2,3, 1e-4))
```
