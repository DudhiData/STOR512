```python
# Bisection Method Example

import math

def func(inter, num):
    return inter ** 2 - num

def bisection(num, tol):
    a = 0
    b = max(1, num)
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if abs(func(c,num)) == 0:
            return c
        if abs(func(c, num)) < tol:
            return c
        elif func(c, num) * func(a, num) < 0:
            b = c
        else:
            a = c
    return c

print(bisection(0.5, 1e-10))
```
